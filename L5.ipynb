```python
#Answer of 1
def is_prime(n):
  if n==1 or n==0:
    return False
  for i in range(2,n):
    if(n%i==0):
      return False
  return True
n = eval(input("Enter a number: "))
for i in range(1,n+1):
      if(is_prime(i)):
        print(i)
```

```python
#Answer of 4
n=eval(input("Enter n: "))
k=eval(input("Enter k: "))
d=(n-k)
def choose(n,k):
  fac1=1
  fac2=1
  fac3=1
  for i in range(1,n+1):
    fac1 =fac1*i
  for j in range(1,k+1):
    fac2=fac2*j
  for s in range(1,d+1):
    fac3=fac3*s
  value=(fac1)//((fac2)*(fac3))
  return value
choose(n,k)



```
